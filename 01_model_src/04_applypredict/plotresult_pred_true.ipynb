```python
import matplotlib.pyplot as plt
import pandas as pd 
import matplotlib
```

```python
df = pd.read_csv("output/randomforest_random_cv_predictions_250918-215814.csv")
```

```python
df_val = df[df['dataset']=='validation'].sample(600)
df_val.shape
```

```python
tx = (80,200)
ty = (80,200)

fig = plt.figure(figsize =(8,8))
plt.plot(df_val['y_true'], df_val["y_pred"],
         linestyle="",
         marker=".",
         markersize = 3,
         color=matplotlib.cm.tab20.colors[6])
ax = plt.gca()
ax.set_xlim(tx[0],tx[1])
ax.set_ylim(ty[0],ty[1])

ax.plot(tx,ty, label = "Best fit")
ax.plot(tx,tuple(y*1.1 for y in ty),
        color = matplotlib.cm.tab20.colors[4],
        ls = '--',
        label="±10%")
ax.plot(tx,tuple(y*0.9 for y in ty),
        color = matplotlib.cm.tab20.colors[4],
        ls = '--')
ax.plot(tx,tuple(y*1.2 for y in ty),
        color = matplotlib.cm.Accent.colors[2],
        ls = '-.',
        label="±20%")
ax.plot(tx,tuple(y*0.8 for y in ty),
        color = matplotlib.cm.Accent.colors[2],
        ls = '-.')

ax.set_ylabel("Alkalinity Predicted (mg/L as CaCO3)")
ax.set_xlabel("Alkalinity Observed (mg/L as CaCO3)")
ax.legend()

plt.show()
fig.savefig("LocAn.png")
```
