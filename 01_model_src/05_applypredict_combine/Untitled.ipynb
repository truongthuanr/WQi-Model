```python
import pandas as pd
from datetime import datetime

# Bước 1: Đọc 2 file CSV
df1 = pd.read_csv("data_4perday_cleaned.csv")
df2 = pd.read_csv("LocAn.csv")

df1['Date'] = pd.to_datetime(df1['Date'], errors='coerce')
df1['Date'] = df1['Date'].dt.strftime('%d/%m/%Y')
df2['Date'] = pd.to_datetime(df2['Date'], errors='coerce')
df2['Date'] = df2['Date'].dt.strftime('%d/%m/%Y')
df2 = df2[df2['Vụ nuôi'].notna()]


# Bước 2: Đảm bảo df2 có đầy đủ các cột giống df1
for col in df1.columns:
    if col not in df2.columns:
        df2[col] = pd.NA  # Thêm cột mới với giá trị null

# Sắp xếp lại thứ tự cột của df2 theo df1
df2 = df2[df1.columns]

# Gộp 2 dataframe
df_combined = pd.concat([df1, df2], ignore_index=True)
    
# Bước 3: Lưu vào file CSV mới
df_combined.to_csv("datacombined.csv", index=False)
```
